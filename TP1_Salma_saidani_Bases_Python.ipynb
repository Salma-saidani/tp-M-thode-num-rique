### 1. Qu'est-ce qu'une variable ?

Une **variable** est comme une boîte qui stocke une valeur. En Python, on crée une variable en lui donnant un nom et en lui assignant une valeur avec le signe `=`.

**Règles pour nommer une variable :**
- Commence par une lettre (a-z, A-Z) ou un underscore (_)
- Peut contenir des lettres, chiffres et underscores
- Pas d'espaces (utilisez _ à la place)
- Sensible à la casse : `Temperature` ≠ `temperature`
- Évitez les mots réservés de Python (if, for, while, etc.)

````python
# Exemples de déclaration de variables
temperature = 25.5
pression = 101325
nom_reacteur = "Réacteur A"
est_actif = True

print("Température :", temperature)
print("Pression :", pression)
````

### 2. Les types de données fondamentaux

Python utilise principalement 4 types de données de base :

| Type | Description | Exemple |
|------|-------------|----------|
| `int` | Nombre entier | `42`, `-15`, `1000` |
| `float` | Nombre décimal | `3.14`, `-0.5`, `2.0` |
| `str` | Chaîne de caractères | `"Bonjour"`, `'Python'` |
| `bool` | Booléen (Vrai/Faux) | `True`, `False` |

Pour connaître le type d'une variable, on utilise la fonction `type()`.

````python
# Exemples de types
nombre_entier = 10
nombre_decimal = 3.14
texte = "Génie des procédés"
vrai_ou_faux = True

print("Type de nombre_entier:", type(nombre_entier))
print("Type de nombre_decimal:", type(nombre_decimal))
print("Type de texte:", type(texte))
print("Type de vrai_ou_faux:", type(vrai_ou_faux))
````

### 3. Opérations arithmétiques

Python permet d'effectuer toutes les opérations mathématiques classiques :

| Opération | Symbole | Exemple | Résultat |
|-----------|---------|---------|----------|
| Addition | `+` | `5 + 3` | `8` |
| Soustraction | `-` | `5 - 3` | `2` |
| Multiplication | `*` | `5 * 3` | `15` |
| Division | `/` | `5 / 2` | `2.5` |
| Division entière | `//` | `5 // 2` | `2` |
| Modulo (reste) | `%` | `5 % 2` | `1` |
| Puissance | `**` | `5 ** 2` | `25` |

````python
# Exemples d'opérations
a = 10
b = 3

print("Addition:", a + b)
print("Soustraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Division entière:", a // b)
print("Modulo:", a % b)
print("Puissance:", a ** b)
````

### 4. Opérations sur les chaînes de caractères

Les chaînes de caractères (strings) se définissent avec des guillemets simples `'...'` ou doubles `"..."`.

**Opérations possibles :**
- **Concaténation** : Assembler deux chaînes avec `+`
- **Répétition** : Répéter une chaîne avec `*`
- **Longueur** : Obtenir le nombre de caractères avec `len()`

````python
# Exemples avec les chaînes
prenom = "salma"
nom = "saidani"

# Concaténation
nom_complet = prenom + " " + nom
print("Nom complet:", nom_complet)

# Répétition
print("Répétition:", "Python " * 3)

# Longueur
print("Longueur du nom:", len(nom_complet))
````

### 5. Conversion de types (Casting)

Parfois, il faut convertir une variable d'un type à un autre. Python propose des fonctions de conversion :

- `int()` : Convertit en entier
- `float()` : Convertit en décimal
- `str()` : Convertit en chaîne de caractères
- `bool()` : Convertit en booléen

````python
# Exemples de conversion
nombre_texte = "42"
nombre_entier = int(nombre_texte)
print("Converti en entier:", nombre_entier, "Type:", type(nombre_entier))

decimal = float("3.14")
print("Converti en float:", decimal, "Type:", type(decimal))

entier = 100
texte = str(entier)
print("Converti en string:", texte, "Type:", type(texte))
````

### 6. Affichage formaté avec print()

Il existe plusieurs façons d'afficher des valeurs avec `print()` :

**Méthode 1 : Séparation par virgules**

````python
temperature = 25.5
print("La température est de", temperature, "°C")
````

**Méthode 2 : f-strings (recommandé)**

````python
temperature = 25.5
pression = 101325
print(f"Température: {temperature}°C, Pression: {pression} Pa")

# Avec formatage (2 décimales)
print(f"Température: {temperature:.2f}°C")
````

### Exercice 10 : Challenge - Calcul complet

Un réacteur chimique fonctionne avec les paramètres suivants :
- Température initiale : 20°C
- Température finale : 85°C
- Masse de réactif : 50 kg
- Capacité calorifique : 4.18 J/(g·°C) pour l'eau

L'énergie nécessaire se calcule avec : **Q = m × c × ΔT**

Où :
- Q = énergie en Joules
- m = masse en grammes
- c = capacité calorifique
- ΔT = variation de température (Tf - Ti)

1. Calculez ΔT
2. Convertissez la masse en grammes
3. Calculez l'énergie nécessaire
4. Convertissez cette énergie en kJ (divisez par 1000)
5. Affichez tous les résultats intermédiaires et le résultat final

````python
Ti = 20
Tf = 85
masse = 50
c = 4.18
delta_T = Tf - Ti
masse_g = masse * 1000
energie = masse_g * c * delta_T
energie_kJ = energie / 1000
print("delta=" , delta_T , "°C")
print("masse en grammes = ", masse_g , "g")
print("energie en Joules =" , energie , "J")
print("energie en kJ=" , energie_kJ , "kJ")
````
